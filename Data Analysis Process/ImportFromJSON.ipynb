```python
import pandas as pd
import requests
```

```python
response = requests.get(
    "https://api.themoviedb.org/3/movie/top_rated?api_key=14cb3aa18123646d9ea2c4ac6100dab7&language=en-US&page=1")
```

```python
df = pd.DataFrame(response.json()["results"])[
    ["id", "title", "overview", "release_date", "popularity", "vote_average", "vote_count"]]
df
```

```python
df = pd.DataFrame()
for i in range(1, 20):
    response = requests.get(
        rf"https://api.themoviedb.org/3/movie/top_rated?api_key=14cb3aa18123646d9ea2c4ac6100dab7&language=en-US&page={i}")
    temp_df = pd.DataFrame(response.json()["results"])[
        ["id", "title", "overview", "release_date", "popularity", "vote_average", "vote_count"]]
    df = pd.concat([df, temp_df], ignore_index=True)
```
